## Release Clause

```python
b['Release Clause'].head(10)
```

```python
#release clause have prefix and sufix € and M which means thousand and million.
#check if all the columns have € and M and change it as 1000 and 1000000.
```

```python
def replace(x):
    x = str(x)
    if '€' in x:
        x = x.replace('€','')
    if "M" in x :
        m  =  10 ** 6
        x  =  x.replace('M','')
        x  =  str(float(x)*m)
    if "K" in x :
        k  =  10 ** 3
        x  =  x.replace('K','')
        x  =  str(float(x)*k)
    return x

b['Release Clause']=b['Release Clause'].apply('replace')
b['Release Clause'].head()
```
